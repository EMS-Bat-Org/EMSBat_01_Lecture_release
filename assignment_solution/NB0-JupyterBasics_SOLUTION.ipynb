### Recursive function example: factorial

Details on the factorial function can be found here, for example: [Factorial](https://de.wikipedia.org/wiki/Fakult%C3%A4t_(Mathematik))

## Local and global variables in functions

```python
def test_loc_glob():
    i = 42 # this variable i is LOCAL to the function test_loc_glob, it is different from the variable i defined outside the function
    s = 'I am a string INSIDE the function.' # this variable s is also LOCAL to the function test_loc_glob, it is different from any variable s defined outside the function
    print(s)
    print(i)

s = 'I am a string OUTSIDE - in the program code.'
i = 7 # this variable i is OUTSIDE the function test_loc_glob, it is different from the variable i defined inside the function

print("Before calling the function:")
test_loc_glob()

print("After calling the function:")    
print(s)
print(i)
```

*** If a variable name defined in a function is to be used both in a function and in the program code, this variable can be defined as a “global” variable by using the keyword `global` followed by the variable name (`i` in the example). Mixing local and global variables with the same name within a function does not work and results in an error. ## Reference or copy?
For lists and DataFrames, it is important to understand whether two variables point to **the same object** or whether a **new copy** was actually created.


### Example: reference vs. copy for lists

- In principle, variables can be passed in two different ways in any programming language: by copy or by reference. A reference can be understood as a pointer to a specific location in memory. In C, variables are generally passed by copy. In that case, only the variable `newList` would be changed in the example, but not the variable `a`.
- In Python, variables are generally handled as references. Therefore, the variables `a` and `newList` are just two pointers to the same memory area. Both pointers can be used to modify the same data. Passing by reference is significantly more performant than passing by copy, especially for larger objects.
- This topic is important because changes made through a reference can **unexpectedly** affect other variables.


```python
# Reference: both variables point to the same list
list_a = ['T', 'a', 'n', 'k']
list_b = list_a
list_b[0] = 'B'

print("list_a:", list_a)
print("list_b:", list_b)
```

If you want to create a copy in Python, you can use the `.copy()` function.

**Remember:** An unintended reference is a common source of errors in data analysis workflows.


```python
# Copy: list_b is a reference, but not a copy!
list_a = ['T', 'a', 'n', 'k']
list_b = list_a.copy()
list_b[0] = 'B'
print("list_a:", list_a)
print("list_b:", list_b)
list_1 = [1, 2, 4, 5, 6]
list_1.insert(2, 3) # insert 3 at index 2 - note that indexing strats with 0 not 1 (!)

print("list_1:", list_1)
list_1.insert(0, 0) # insert 0 at index 0
print("list_1:", list_1)
```

## Pandas DataFrames for time series

A **DataFrame** is like an Excel table, but controllable through code.

In *MATLAB*, a `table` has a similar function (e.g. `T = table([1;2;3], ["A";"B";"C"], [true;false;true], 'VariableNames', {'ID', 'Name', 'Valid'})`).

In this course, we use DataFrames to process *time series* such as:

- load (kW)
- PV power (kW)
- electricity price (€/kWh)
- temperature (°C)


```python
# Create a small DataFrame from a dictionary
sensor_data = {
    "sensor": ["S1", "S2", "S3"],
    "voltage_v": [401.2, 398.7, 403.1],
    "current_a": [15.2, 14.8, 15.5],
}

sensor_df = pd.DataFrame(sensor_data)

```

```python
sensor_df
```

We now define a second DataFrame `df`. It is a **synthetic** dataset, so everything runs without an external file. It contains four time-series signals named `p_load_kw`, `p_pv_kw`, `price_eur_per_kwh`, and `temperature_c`.

```python
# 365 days, 15-min resolution
idx = pd.date_range("2025-01-01", periods=365*24*4, freq="15min")

# time vectors
t_day = (idx.hour + idx.minute/60) / 24          # fraction of day [0..1]
t_year = np.linspace(0, 2*np.pi, len(idx))       # slow annual variation

# PV: DAILY sine pattern (0 at night, peak at noon)
pv_daily = np.sin(2*np.pi*t_day - np.pi/2)
pv_kw = np.clip(3.0 * pv_daily, 0, None)

# Load: daily variation + small annual variation + noise
load_kw = (
    2.0
    + 0.6*np.sin(2*np.pi*t_day)                 # daily oscillation
    + 0.3*np.sin(t_year)                        # slow yearly modulation
    + 0.2*np.random.randn(len(idx))             # noise
)

# Price: slower oscillation + noise
price = (
    0.25
    + 0.05*np.sin(t_year)                       # annual cycle
    + 0.02*np.random.randn(len(idx))
)

# Temperature: daily and seasonal variation
temperature_c = (
    12
    + 8*np.sin(t_year - 0.8)
    + 4*np.sin(2*np.pi*t_day - np.pi/3)
    + 0.8*np.random.randn(len(idx))
)

df = pd.DataFrame({
    "p_load_kw": load_kw,
    "p_pv_kw": pv_kw,
    "price_eur_per_kwh": price,
    "temperature_c": temperature_c,
}, index=idx)
```

### 4.2 Missing values (NaN)

Real-world data often has gaps. Typical strategies are:

- `dropna()` (remove) — often not ideal for time series
- `ffill()` (*forward fill*) — propagate the previous valid value forward
- `interpolate()` — estimate intermediate values


### Plotly example for aggregated data

Aggregated results such as daily mean values or daily maxima should also be visualized briefly.


```python
daily_plot_df = daily_price.reset_index()

fig = px.line(
    daily_plot_df,
    x="index",
    y=["min", "max", "mean"],
    title="Daily electricity price statistics",
    labels={"index": "Day", "value": "Price [€/kWh]", "variable": "Metric"},
)
fig.update_layout(legend_title_text="Metric")
fig.show()
```

```python
# Demo: create NaNs artificially
df_nan = df.copy()
df_nan.iloc[100:120, df_nan.columns.get_loc("p_load_kw")] = np.nan

print("NaNs in p_load_kw:", df_nan["p_load_kw"].isna().sum())
```
